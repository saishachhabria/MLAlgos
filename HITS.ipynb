```python
import numpy as np

n = int(input("Enter number of nodes: "))
graph = []
for i in range(n):
    graph.append([int(x) for x in input().split(" ")])
#graph = [[0, 1, 1, 1], [0, 0, 0, 1], [1, 0, 0, 1], [1, 0, 1, 0]]
hub = []
auth = []

for i in range(n):
    hub.append(1)
    auth.append(1)
k = 0
while(k < 8):
    newauth = []
    newhub = []
    for i in range(n):
        newauth.append(0)
        newhub.append(0)
    for i in range(n):
        for j in range(n):
            if graph[j][i] == 1:
                newauth[i] += hub[j]
            if graph[i][j] == 1:
                newhub[i] += auth[j]
    sumhub = sum(newhub)
    sumauth = sum(newauth)
    for i in range(n):
        newauth[i] /= sumauth
        newhub[i] /= sumhub
    auth = newauth
    hub = newhub
    print("\nIteration",k,":\nHub\t\t\tAuthority")
    for i in range(n):
        print(newhub[i],"\t",newauth[i])
    k += 1
```
