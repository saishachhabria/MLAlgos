```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

data = pd.read_csv("final.csv")
data.columns = ['Street','Neighborhood','HouseStyle','YrSold','SalePrice','Class']

#x = list(data.SalePrice)
#y = list(data.YrSold)

x = [0.40,0.22,0.35,0.26,0.08,0.45]
y = [0.53,0.38,0.32,0.19,0.41,0.30]
dataset = [(x[i],y[i]) for i in range(len(x))]

x = dataset
matrix = []
items = []

for i in range(len(dataset)):
    matrix.append([])
    for j in range(len(dataset)):
        dist = round(math.sqrt((x[i][0]-x[j][0])**2 + (x[i][1]-x[j][1])**2),2)
        matrix[i].append(dist)

def print_matrix():
    print("Distance Matrix:\n")
    for i in range(len(items)):
        print("\t\t",str(items[i]),end="")
    print("\n")
    for i in range(len(matrix)):
        print(items[i],end="\t\t")
        for j in range(len(matrix[i])):
            if i > j:
                print(matrix[i][j],end="\t\t")
        print()

def find_min():
    min = 99999
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[j][i] < min and i != j:
                min = matrix[i][j]
                #y1 = [items[i]]
                #y2 = [items[j]]
                yy = (items[i],items[j])
    return yy


for i in range(len(matrix)):
    items.append(i)
print_matrix() 

while(len(items) != 1):   
    (m,n) = find_min()
    print("\nMinimum distance pair: (",m,",",n,")")
    print("\nForming a cluster, we get: ")
    k = (m,n)
    m = items.index(m)
    n = items.index(n)
    for j in range(len(matrix)):
        if matrix[j][m] < matrix[j][n]:
            min = matrix[j][m]
        else:
            min = matrix[j][n]
        matrix[j][n] = min
        matrix[n][j] = min
    items[n] = k
    for i in range(len(matrix)):
        matrix[i].pop(m)
    matrix.pop(m)
    items.pop(m)
    print_matrix() 
```
