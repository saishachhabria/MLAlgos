```python
import numpy as np
import math

lines = []
calc_yes = calc_no = 0
with open("test1.csv",'r') as fp:
    lines = fp.readlines()

def getUniqueValues(attr):
    vals = []
    index = getIndex(attr)
    for i in range(len(dataset)):
        if dataset[i][index] not in vals:
            vals.append(dataset[i][index])
    return vals

def getIndex(attr):
    for i in fields:
        if i == attr:
            return fields.index(i)
    return -1

def calc(val,attr):
    count = 0
    index = getIndex(attr)
    for i in range(len(dataset)):
        if dataset[i][index] == val:
            count += 1
    return count

def calcAttr(val,attr):
    count1 = 0
    count2 = 0
    index = getIndex(attr)
    for i in range(len(dataset)):
        if dataset[i][index] == val:
            if dataset[i][4] == "Yes":
                count1 += 1
            else:
                count2 += 1
    return [count1,count2]

def calcInfoGain(p,n):
    t1 = p/(p+n)
    t2 = n/(p+n)
    if t1 == 0:
        return (-(t2*math.log(t2,2)),4)
    elif t2 == 0:
        return round((-t1*math.log(t1,2)),4)
    else:
        return round(((-t1*math.log(t1,2))-(t2*math.log(t2,2))),4)

dataset = []
fields = []
firstLine = True
for line in lines:
    if firstLine == True:
        firstLine = False
        fields = line.split(",")
    else:
        line = line.rstrip()
        dataset.append(line.split(","))

calc_yes = calc("Yes","Class")
calc_no = calc("No","Class")
infoGain = calcInfoGain(calc_yes,calc_no)
print("\nP[Yes] =",calc_yes)
print("P[No] = ",calc_no)
print("Information Gain =",infoGain,"\n")
entropy = []
maxGain = 0
attrib = ""
for attribute in fields[:4]:
    info = []
    print("\nAttribute",attribute,":")
    vals = getUniqueValues(attribute)
    for i in vals:
        counts = calcAttr(i,attribute)
        print(i," - ",counts)
        info.append([(counts[0]+counts[1]),calcInfoGain(counts[0],counts[1])])
    entropy = 0
    for i in info:
        entropy += i[0]/len(dataset)*i[1]
    entropy = infoGain - entropy
    if entropy > maxGain:
        maxGain = entropy
        attrib = attribute
    print(info)   
    print("Entropy = ",entropy,"\n")

print("\nMaximum gain attribute (root of the tree): ",attrib)
```
