```python
import pandas as pd
import numpy as np

def generateC1(dataset):
    candidateSet = {}
    data = []
    for i in dataset:
        for product in i:
            if product not in candidateSet:
                candidateSet[product] = 1
            else:
                candidateSet[product] = candidateSet[product] + 1
    for key in candidateSet:
        data.append([key,candidateSet[key]])
    return data

def higherThanMin(dataset,minSupport):
    newdata = []
    for i in dataset:
        if i[1] >= minSupport:
            newdata.append(i)
    return newdata

def genCombinations(dataset,count):
    combinationArray = []
    data = []
    for i in dataset:
        for j in dataset:
            comb = [k for k in i]
            for k in j:
                if k not in comb:
                    comb.append(k)
            comb.sort()
            if comb not in combinationArray and i!=j and len(comb)==count:
                combinationArray.append(comb)
    return combinationArray

def getFrequency(dataset,comb):
    newcount = []  
    for j in comb:
        count = 0
        for i in dataset:
            flag = True         
            for k in j:
                if k not in i:
                    flag = False
            if flag == True:
                count = count + 1
        newcount.append([j,count])
    return newcount

def getCandidateList(data):
    items = []
    for i in data:
        try:
            someobject = iter(i[0])
            items.append(i[0])
        except TypeError:
            items.append([i[0]])
    return items

def getRules(data):
    rules = []
    for i in data:
        for j in i:
            lhs = []
            rhs = []
            lhs.append(j)
            for k in i:
                if k not in lhs:
                    rhs.append(k)
            rules.append([lhs,rhs])
            rules.append([rhs,lhs])
    return rules
    

    
    

data = pd.read_csv("final.csv",header=None)

dataset = []
firstLine = True
with open("final.csv",'r') as fp:
    lines = fp.readlines()
    
for line in lines:
    if firstLine == True:
        firstLine = False
    else:    
        line = line.rstrip()
        dataset.append(line.split(","))
        
        
print("Association Rule Mining (Apriori)\n")
dataset = [[1,2,4],[1,2,5],[1,3,5],[2,4],[2,3],[1,2,3,5],[1,3],[1,2,3],[2,3],[3,5]]
#dataset = [['Bread','Jelly','Butter'],['Bread','Butter'],['Bread','Milk','Butter'],['Coke','Bread'],['Coke','Milk']]
noOfTransactions = len(dataset)
minSupport = int(int(input("\nEnter minimum support: "))*noOfTransactions/100)
minConfidence = int(input("\nEnter minimum confidence: "))

print("\n\nInitialization(iteration 1):")
candidateSet1 = generateC1(dataset)
print("Candidate set: ",candidateSet1)
frequentSet1 = higherThanMin(candidateSet1,minSupport)
print("Frequent set: ",frequentSet1,"\n")

items = getCandidateList(frequentSet1)
count = 2
while(count <= 3):
    print("\nIteration",count,":")
    comb = genCombinations(items,count)
    print("Combinations: ",comb)
    freq = getFrequency(dataset,comb)
    print("Frequency of set: ",freq)
    freqHigher = higherThanMin(freq,minSupport)
    print("Frequency above threshold: ",freqHigher,end="\n\n")
    count = count + 1  
    items = getCandidateList(freqHigher)

if len(freqHigher) == 0:
    print("No such rules!")
else:
    try:
        someobject = iter(items[0])
        aprioriRules = getRules(items)
    except:
        aprioriRules = getRules([items[0]])

    print("Rules\t\tSupport\tConfidence\tConfidence %")
    for i in aprioriRules:
        rule = ""
        vals = []
        for j in i[0]:
            rule += str(j)+" ^ "
            vals.append(j)
        count = getFrequency(dataset,[vals])
        rule = rule[0:len(rule)-3]
        rule += " --> "
        for j in i[1]:
            rule += str(j)+" ^ "
            vals.append(j)
        rule = rule[0:len(rule)-3]
        print(rule,end="\t")
        vals.sort()
        conf = 0
        for j in freqHigher:
            if vals == j[0]:
                print(j[1],"\t",j[1],"/",count[0][1],end="\t\t")
                conf = round(j[1]/count[0][1]*100,2)
        print(conf,"\t",end="\t")
        if conf > minConfidence:
            print("ACCEPTED RULE")
        else:
            print()
```
